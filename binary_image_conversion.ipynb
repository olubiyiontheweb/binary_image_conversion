```python
#import libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("zebra.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img,cmap='gray')
```

```python
thresh = 125 # allow colors below 150
max = 255 # set 255 for all colors more than thresh
threshed, dst = cv2.threshold(img,thresh,max,cv2.THRESH_BINARY)
```

```python
plt.imshow(dst, cmap='gray')
```
